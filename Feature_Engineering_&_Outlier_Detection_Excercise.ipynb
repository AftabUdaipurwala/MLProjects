```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
```

```python
df= pd.read_csv('/content/drive/MyDrive/py-master/ML/FeatureEngineering/1_outliers/Exercise/AB_NYC_2019.csv')
df.head(4)
```

```python
plt.hist(df.price) # definately this has a lot of outliers
```

```python
plt.boxplot(df.price)
```

```python
min_threshold, max_threshold = df.price.quantile([0.01,0.95])
```

```python
df2= df[(df['price']<max_threshold) & (df['price']>min_threshold)] # Cleaned Dataset
```

```python
plt.hist(df2.price) # though the data is skewed but outliers are very few
```

```python
plt.boxplot(df2.price) # there are very few outliers now
```
